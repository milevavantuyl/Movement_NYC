```python
# read 2015 data csv file into a dataframe

csv_file = "~/IDC5131_Dist_Comp/Project_2_files/exported_csvs/2015/uber_janjune15_wcodes_centroids.csv"
uber2015 = pd.read_csv(csv_file)
uber2015.head()
```

```python
# create column headings
col_order = ['datetime', 'lat', 'lon', 'base', 'locationID', 'borough', 'zone']
```

```python
# reformat 2015 data
# Convert date to datetime format
uber2015['datetime'] = pd.to_datetime(uber2015['Pickup_date'])

# Drop extraneous columns
uber2015 = uber2015.drop(['Unnamed: 0', 'Dispatching_base_num', 'Pickup_date'], axis = 1)

# Rename columns to obtain the same format
uber2015 = uber2015.rename(columns = {"Affiliated_base_num": "base", "xCentroid": "lon", "yCentroid":"lat"})

# Reorder the columns
uber2015 = uber2015[col_order]

# Sort by datetime
uber2015 = uber2015.sort_values("datetime").reset_index(drop = True)

# Final format
uber2015.head()
```

```python
# split 2015 data into separate data frames by month

jan2015 = uber2015[uber2015['datetime'].dt.date < pd.to_datetime('2015-02-01').date()]
feb2015 = uber2015[(uber2015['datetime'].dt.date > pd.to_datetime('2015-01-31').date()) & (uber2015['datetime'].dt.date < pd.to_datetime('2015-03-01').date())]
march2015 = uber2015[(uber2015['datetime'].dt.date > pd.to_datetime('2015-02-28').date()) & (uber2015['datetime'].dt.date < pd.to_datetime('2015-04-01').date())]
april2015 = uber2015[(uber2015['datetime'].dt.date > pd.to_datetime('2015-03-31').date()) & (uber2015['datetime'].dt.date < pd.to_datetime('2015-05-01').date())]
may2015 = uber2015[(uber2015['datetime'].dt.date > pd.to_datetime('2015-04-30').date()) & (uber2015['datetime'].dt.date < pd.to_datetime('2015-06-01').date())]
june2015 = uber2015[uber2015['datetime'].dt.date > pd.to_datetime('2015-05-31').date()]
```

```python
# load all dataframes into list that can be iterated over

df_list = [april2014, may2014, june2014, july2014, august2014, september2014, jan2015, feb2015, march2015, april2015, may2015, june2015]
```

```python
# for every dataframe in list, add a counter column of ones, then group by locationID, borough, and zone, and sum the counts

new_data = []
for i in df_list:
    i['count'] = 1
    j = i.groupby(['locationID', 'borough', 'zone']).sum('count').reset_index()
    new_data.append(j)
```

```python
# create a list of corrective coefficients to account for monthly growth in uber patronage over the period of analysis  
# this will allow us to make apples to apples comparisons for departures between months

sum_counts = []
for i in new_data:
    sum_counts.append(i['count'].sum())
    
coefficients = np.mean(sum_counts)/sum_counts
coefficients
```

```python
# add empty column, corrected count, to each dataframe in list

for i in new_data:
    i["corrected_count"] = None
```

```python
# fill empty column with corrected counts (count * corrective coefficient for each month)

for i,j in zip(new_data, coefficients):
    i['corrected_count'] = i['count'] * j
```

```python
# load taxi zones as data frames and subset those zones with index locationID < 264

taxi_zones = pd.read_csv("/home/richard/Downloads/taxi+_zone_lookup.csv")
taxi_zones = taxi_zones.rename(columns = {"LocationID": "locationID"})
taxi_zones = taxi_zones[taxi_zones.locationID < 264]
```

```python
# find missing locationIDs in all data frames, insert those location IDs and fill in other elements in row with 0
# add all ride counts from the dfs into a list

all_counts = []
for count,df in enumerate(new_data):
    diff = pd.DataFrame(taxi_zones.locationID[~taxi_zones.locationID.isin(df.locationID)])
    join_diff = pd.concat([df,diff], axis = 0, ignore_index=True)
    new_data[count] = pd.concat([new_data[count],diff], axis = 0, ignore_index=True)
    join_diff = join_diff.fillna(0)
    new_data[count] = new_data[count].fillna(0)
    join_diff = join_diff.sort_values(by='locationID', ignore_index = True)
    new_data[count] = join_diff.sort_values(by='locationID', ignore_index = True)
    all_counts.append(join_diff['count'])
```

```python
# calculate average ridership per location_ID and insert into list

avg_counts = np.mean(all_counts, axis = 0).astype(float)
```

```python
# add empty column, rider shifts (denoting shifts from average), to each dataframe in list

for df in new_data:
    df["rider_shift"] = None
```

```python
# fill empty column with corrected counts (count * corrective coefficient for each month)

for df in new_data:
    df['rider_shift'] = df['corrected_count'] - avg_counts
```

```python
# create list of csv titles
df_list_strs = ['april2014', 'may2014', 'june2014', 'july2014', 'august2014', 'september2014', 'jan2015', 'feb2015', 'march2015', 'april2015', 'may2015', 'june2015']
```

```python
# export every df in new_data list as csv file

for count, df in enumerate(new_data):
    df.to_csv("~/IDC5131_Dist_Comp/Project_2_files/csvs_w_vars/" + df_list_strs[count] + ".csv")
```
